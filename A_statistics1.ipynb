````python
# Statistics Assignment

## 1. Explain the properties of the F-distribution.
The F-distribution is a continuous probability distribution that arises frequently in variance analysis. It has the following properties:
- It is positively skewed and only takes positive values.
- It is defined by two degrees of freedom parameters (df1, df2).
- The mean of the F-distribution is approximately **df2 / (df2 - 2)** for df2 > 2.
- As the degrees of freedom increase, the F-distribution approximates the normal distribution.
- It is commonly used in hypothesis testing, such as ANOVA and variance ratio tests.

## 2. In which types of statistical tests is the F-distribution used, and why is it appropriate for these tests?
The F-distribution is primarily used in:
- **ANOVA (Analysis of Variance)**: To compare means of multiple groups by analyzing variance.
- **F-test for Equality of Variances**: To compare variances of two populations.
- **Regression Analysis**: To determine the significance of regression models.
It is appropriate for these tests because it models the ratio of variances, which helps in determining if observed differences are statistically significant.

## 3. What are the key assumptions required for conducting an F-test to compare the variances of two populations?
To perform an F-test, the following assumptions must be met:
- The samples are **independent** of each other.
- The data is **normally distributed** in both populations.
- The samples are drawn **randomly** from their respective populations.
- The populations have **equal variances**, unless testing for variance equality.

## 4. What is the purpose of ANOVA, and how does it differ from a t-test?
ANOVA (Analysis of Variance) is used to compare **more than two group means** to determine if at least one group significantly differs. A t-test, on the other hand, is used for comparing the means of **only two groups**. The key difference:
- **t-test**: Compares two groups' means.
- **ANOVA**: Compares multiple groups' means simultaneously, reducing Type I error.

## 5. Explain when and why you would use a one-way ANOVA instead of multiple t-tests when comparing more than two groups.
A one-way ANOVA is preferred over multiple t-tests because:
- **Prevents Type I error inflation**: Running multiple t-tests increases the likelihood of false positives.
- **More efficient**: ANOVA tests all group differences in one test.
- **More statistical power**: Detects group differences better than separate t-tests.

## 6. Explain how variance is partitioned in ANOVA into between-group variance and within-group variance. How does this partitioning contribute to the calculation of the F-statistic?
- **Between-group variance**: Variability due to differences between group means.
- **Within-group variance**: Variability within each group due to random error.
- **F-statistic Calculation**:
  \[ F = \frac{\text{Between-group variance}}{\text{Within-group variance}} \]
A high F-value suggests significant differences among groups.

## 7. Compare the classical (frequentist) approach to ANOVA with the Bayesian approach. What are the key differences in terms of how they handle uncertainty, parameter estimation, and hypothesis testing?
| Aspect | Classical ANOVA | Bayesian ANOVA |
|--------|----------------|----------------|
| **Uncertainty** | Uses p-values to determine significance | Uses probability distributions |
| **Parameter Estimation** | Estimates mean differences using sample data | Uses prior distributions and updates beliefs |
| **Hypothesis Testing** | Null hypothesis rejection based on p-value | Probability of model correctness based on prior and posterior beliefs |

## 8. Perform an F-test to determine if the variances of two professions' incomes are equal.
### Given Data:
- **Profession A**: [48, 52, 55, 60, 62]
- **Profession B**: [45, 50, 55, 52, 47]

### Python Code:
```python
import scipy.stats as stats
import numpy as np

# Data
profession_A = [48, 52, 55, 60, 62]
profession_B = [45, 50, 55, 52, 47]

# Variance
var_A = np.var(profession_A, ddof=1)
var_B = np.var(profession_B, ddof=1)

# F-statistic
F = var_A / var_B
p_value = stats.f.cdf(F, len(profession_A)-1, len(profession_B)-1)

print("F-statistic:", F)
print("P-value:", p_value)
```
### Conclusion:
If p-value < 0.05, reject null hypothesis (variances are different); otherwise, assume equal variances.

## 9. Conduct a one-way ANOVA to test whether there are statistically significant differences in average heights between three regions.
### Given Data:
- **Region A**: [160, 162, 165, 158, 164]
- **Region B**: [172, 175, 170, 168, 174]
- **Region C**: [180, 182, 179, 185, 183]

### Python Code:
```python
import scipy.stats as stats

# Data
region_A = [160, 162, 165, 158, 164]
region_B = [172, 175, 170, 168, 174]
region_C = [180, 182, 179, 185, 183]

# One-way ANOVA
F_stat, p_value = stats.f_oneway(region_A, region_B, region_C)

print("F-statistic:", F_stat)
print("P-value:", p_value)
```
````
